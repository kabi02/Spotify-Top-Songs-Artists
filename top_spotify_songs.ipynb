# Data Exploration and Cleaning

Since the data that we have is too large for Excel to handle, we'll do it in Python instead using the Pandas library.

```python
# importing the pandas library to the script

import pandas as pd
```

```python
# loading the dataset into a dataframe

df = pd.read_csv('..\\..\\Datasets\\top_spotify_songs.csv')
```

```python
# checking for missing values

df.isna().sum()
```

```python
# first we handle the missing values in the 'name' column

missing_name = df[df['name'].isna()==1]
```

```python
list = missing_name.spotify_id.unique()
print(list)
```

These are all the IDs of songs which lack titles

```python
# we try to search for the missing titles from the original dataframe

filtered_df = df[df['spotify_id'].isin(list)]
filtered_df
```

```python
filtered_df = filtered_df[filtered_df['name'].isna()==0].drop_duplicates(subset='spotify_id')
filtered_df
```

I filtered the dataframe even more to see only those which actually have titles in the original data

```python
# we make a dictionary for the song IDs and their corresponding titles

dict = {
    "0y2sUHVDAWQbfuIQGEfhFO" : "NO POLICE (Remix)",
    "2V2K1hzCtgj9xAnga9WUTy" : "Horns In The Sun - Thakzin Remix",
    "6yxtsR3nc3aUL1wcbLn8A3" : "Jaga Jaga"
}

dict
```

```python
# now we map those titles into our temporary 'missing_name' dataframe

missing_name['name'] = missing_name['spotify_id'].map(dict)
missing_name
```

```python
dict = {
    "0y2sUHVDAWQbfuIQGEfhFO" : "CynMiah, Magnito",
    "2V2K1hzCtgj9xAnga9WUTy" : "DJ Kent, Thakzin, Mo-T, MÖRDA, Brenden Praise",
    "6yxtsR3nc3aUL1wcbLn8A3" : "VicK, Babi"
}
```

```python
# and mapped them into their respective rows

missing_name['artists'] = missing_name['spotify_id'].map(dict)
missing_name
```

```python
missing_name = missing_name.dropna(subset=['name'])
```

Since there's no way for us to retrieve the original information on what these songs may be, I just decided to not include them in the final cleaned data.

** If we really want to find information regarding these songs, a possible solution for is to use the Spotify API to try and look for the using using their IDs but that is outside the scope for this project.

```python
# now we update the original dataframe with all the cleaning that we did on our 'missing_name' dataframe

df.update(missing_name)
df
```

```python
# we also dropped the rows that had missing values in the 'name' column in the original dataframe

df = df.dropna(subset="name")
df.isna().sum()
```

```python
country_df = df[df['country'].isna()==1]
country_df
```

```python
# for the country, I just filled the missing values with the 'Unknown' value

country_df["country"] = country_df.country.fillna('Unknown')
```

```python
country_df
```

```python
country_df.isna().sum()
```

```python
df.update(country_df)
df.isna().sum()
```

```python
df.columns
```

### I should have done this right at the start, but here I dropped the columns that I'm not going to use for the visualizations

```python
cols_to_drop = ['daily_movement', 'weekly_movement', 'is_explicit', 'duration_ms', 'danceability', 'energy', 'key', 'loudness', 'mode', 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence', 'tempo', 'time_signature']
df.drop(cols_to_drop, axis=1, inplace=True)
df
```

```python
# same things here, I just marked entries with missing album names as 'Unknown'

temp = df[df['album_name'].isna()==1]
temp.album_name = temp.album_name.fillna('Unknown')
df.update(temp)
df.isna().sum()
```

```python
# here I changed the data type of the date columns to the proper data type so that they work properly for the visualization that we're making

temp = df[['snapshot_date', 'album_release_date']] = df[['snapshot_date', 'album_release_date']].apply(pd.to_datetime, errors='coerce')
temp.dtypes
```

```python
# finally we update the original dataframe with the changes we made

df.update(temp)
df.dtypes
```
